```python
#part 1 
#1.1. Create two 2D arrays representing coordinates x, y on the cartesian plan
#Both should cover the range (-2, 2) and have 100 points in each direction
```

```python
x = np.linspace(-2,2,100)
y= np.linspace(-2,2,100)
```

```python
#1.2 Visualize each 2D array using pcolormesh
#Use the correct coordiantes for the x and y axes.
```

```python
xx,yy = np.meshgrid(x,y) 
```

```python
plt.pcolormesh(xx)
```

```python
plt.pcolormesh(yy)
```

```python
#1.3 From your cartesian coordinates, create polar coordinates 
#Refer to the wikipedia page for the conversion formula. You will need to use numpy’s arctan2 function. Read its documentation.
```

```python
r = np.sqrt(xx**2 + yy**2)
O = np.arctan2(yy,xx)
```

```python
#1.4 Visualize r and o on the 2d plane using pcolormesh
```

```python
plt.pcolormesh(xx,yy,r)
```

```python
plt.pcolormesh(xx,yy,O)
```

```python
#1.5 Caclulate the quanity And plot it on the x
#y$ plane
```

```python
f = np.cos(4*r)**2 + np.sin(4*O)**2
```

```python
plt.pcolormesh(xx,yy,f)
```

```python
#1.6 Plot the mean of f with respect to the x axis
#as a function of y
```

```python
f_xmean = f.mean(axis=1)
plt.plot(f_xmean, y);
```

```python
#1.7 Plot the mean of f with respect to the y axis
#as a function of x
```

```python
f_ymean = f.mean(axis=0)
plt.plot(x, f_ymean);
```

```python
#2.6 Account For Missing Data
#The profiles contain many missing values. These are indicated by the special “Not a Number” value, or np.nan.
#When you take the mean or standard deviation of data with NaNs in it, the entire result becomes NaN. 
#Instead, if you use the special functions np.nanmean and np.nanstd, you tell NumPy to ignore the NaNs.
#Recalculate the means and standard deviations as in the previous sections using these functions and plot the results.
```

```python
newmeanT = np.nanmean(T, axis = 1)
newstdT = np.nanstd(T, axis = 1)
plt.plot(newmeanT, level)
plt.errorbar(newmeanT, level, xerr = newstdT)
plt.grid(True, c = 'black')
plt.gca().invert_yaxis()
```

```python
newmeanS = np.nanmean(S, axis = 1)
newstdS = np.nanstd(S, axis = 1)
plt.plot(newmeanS, level)
plt.errorbar(newmeanS, level, xerr = newstdS)
plt.grid(True, c = 'black')
plt.gca().invert_yaxis()
```

```python
newmeanP = np.nanmean(P, axis = 1)
newstdP = np.nanstd(P, axis = 1)
plt.plot(newmeanP, level)
plt.errorbar(newmeanP, level, xerr = newstdP)
plt.grid(True, c = 'black')
plt.gca().invert_yaxis()
```
