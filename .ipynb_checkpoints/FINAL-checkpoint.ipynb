```python
import numpy as np 
from matplotlib import pyplot as plt
import pandas as pd
from scipy import optimize 
from scipy import stats 
```

```python
df = pd.read_csv("temporarydata.csv")
```

```python
fig, axes = plt.subplots(1,1, figsize =(20,15))
axes.plot(df['Date'], df['Daily Average Temperature (F)'], label = "Daily Average Temperature (F)", color = "black")
axes.plot(df['Date'], df['Daily Maximum Temperature (F)'], label = "Daily Average Temperature (F)", color = "red")
axes.plot(df['Date'], df['Daily Minimum Temperature (F)'], label = "Daily Average Temperature (F)", color = "blue")
axes.set_title("Daily Average Temperature over Time in Milwaukee, Wisconsin")
axes.set_xlabel("Date")
axes.set_ylabel("Temperature (°F)")
axes.legend(loc="best")
plt.show()
```

```python
def seasonalvariation(x, a, b, c):
    P = 365.25
    return a * np.cos((2 * np.pi / P) * x) + b * np.sin((2 * np.pi / P) * x) + c
df['Date'] = pd.to_datetime(df.iloc[:, 0], format='%m/%d/%y', errors='coerce')
time = (df['Date'] - df['Date'].iloc[0]).dt.days
temperature = pd.to_numeric(df.iloc[:, 1], errors='coerce')

p, covariance = optimize.curve_fit(seasonalvariation, xdata=time, ydata=temperature)
a, b, c = p
predicted_temperature_seasonal = seasonalvariation(time, a, b, c)
error_seasonal = temperature - predicted_temperature_seasonal

start_date = df['Date'].iloc[0]
years = (df['Date'] - start_date).dt.days / 365.25 + 2000  # Convert days to years

fig, ax = plt.subplots(3, 1, figsize=(20, 15), sharex=True)
ax[0].plot(years, temperature, label='Average Daily Temperature (°F)', color='blue')
ax[0].set_title('Average Daily Temperature (°F)')
ax[0].set_ylabel('Temperature (°F)')
ax[0].set_xlabel('Year')
ax[0].legend()
ax[0].grid(True)
ax[1].plot(years, temperature, label='Average Daily Temperature (°F)', color='blue')
ax[1].set_title('Average Daily Temperature (°F)')
ax[1].set_ylabel('Temperature (°F)')
ax[1].set_xlabel('Year')
ax[1].legend()
ax[1].grid(True)
ax[1].plot(years, predicted_temperature_seasonal, label='Predicted Temperature due to Seasonal Variation', color='red')
ax[1].set_xticks(np.linspace(min(years), max(years), num=10))  # help with this
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))
ax[2].plot(years, error_seasonal, label='Prediction Error (Residuals)', color='green')
ax[2].set_title('Prediction Error vs. Fourier Model (Seasonal Residuals)')
ax[2].set_ylabel('Error (°F)')
ax[1].set_xlabel('Year')
ax[2].legend()
ax[2].grid(True)
plt.show()
```

```python
#creating a better model of the prediction error
def fourier_series(x, a0, a1, b1, a2, b2, a3, b3, a4, b4, a5, b5, a6, b6, a7, b7, a8, b8, a9, b9, a10, b10):
    period_1 = 365.25 
    period_2 = 365.25 * 2 
    period_3 = 365.25 * 3
    period_4 = 365.25 * 4
    period_5 = 365.25 * 5
    period_6 = 365.25 * 6
    period_7 = 365.25 * 7
    period_8 = 365.25 * 8
    period_9 = 365.25 * 9
    period_10 = 365.25 * 10
    
    return a0 + a1 * np.cos(2 * np.pi * x / period_1) + b1 * np.sin(2 * np.pi * x / period_1) + a2 * np.cos(2 * np.pi * x / period_2) + b2 * np.sin(2 * np.pi * x / period_2) + a3 * np.cos(2 * np.pi * x / period_3) + b3 * np.sin(2 * np.pi * x / period_3) + a4 * np.cos(2 * np.pi * x / period_4) + b4 * np.sin(2 * np.pi * x / period_4) + a5 * np.cos(2 * np.pi * x / period_5) + b5 * np.sin(2 * np.pi * x / period_5)+a6 * np.cos(2 * np.pi * x / period_6) + b6 * np.sin(2 * np.pi * x / period_6)+a7 * np.cos(2 * np.pi * x / period_7) + b7 * np.sin(2 * np.pi * x / period_7)+a8 * np.cos(2 * np.pi * x / period_8) + b8 * np.sin(2 * np.pi * x / period_8)+a9 * np.cos(2 * np.pi * x / period_9) + b9 * np.sin(2 * np.pi * x / period_9)+a10 * np.cos(2 * np.pi * x / period_10) + b10 * np.sin(2 * np.pi * x / period_10)
p_fourier, _ = optimize.curve_fit(fourier_series, xdata=time, ydata=error_seasonal) #got help with this
a0, a1, b1, a2, b2, a3, b3, a4, b4, a5, b5, a6, b6, a7, b7, a8, b8, a9, b9, a10, b10 = p_fourier
predicted_error_fourier = fourier_series(time, a0, a1, b1, a2, b2, a3, b3, a4, b4, a5, b5, a6, b6, a7, b7, a8, b8, a9, b9, a10, b10)
error_fourier = error_seasonal - predicted_error_fourier  

def fourier_series_basic(x, a0, a1, b1, a2, b2):
    period_1 = 365.25 
    period_2 = 365.25 * 2 
    return a0 + a1 * np.cos(2 * np.pi * x / period_1) + b1 * np.sin(2 * np.pi * x / period_1) + a2 * np.cos(2 * np.pi * x / period_2) + b2 * np.sin(2 * np.pi * x / period_2)
p_fourier_basic, _ = optimize.curve_fit(fourier_series_basic, xdata=time, ydata=error_seasonal) #got help with this
a0, a1, b1, a2, b2 = p_fourier_basic
predicted_error_fourier_basic = fourier_series_basic(time, a0, a1, b1, a2, b2)
error_fourier_basic = error_seasonal - predicted_error_fourier_basic  

fig, ax = plt.subplots(3,1,figsize=(20,10))
ax[0].plot(years, temperature, label='Average Daily Temperature (°F)', color='blue')
ax[0].plot(years, predicted_temperature_seasonal, label='Predicted Seasonal Variation', color='red')
ax[0].set_title('Average Daily Temperature (°F)')
ax[0].set_ylabel('Temperature (°F)')
ax[0].legend()
ax[0].grid(True)
ax[0].plot(years, predicted_temperature_seasonal, label='Predicted Temperature due to Seasonal Variation', color='red')
ax[0].set_xticks(np.linspace(min(years), max(years), num=10)) #I had help with this
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))
ax[1].plot(years, error_seasonal, label='Prediction Error (Residuals)', color='green')
ax[1].plot(years, predicted_error_fourier, label='Enhanced Predicted Periodic Temperature Model', color='red')
ax[1].plot(years, predicted_error_fourier_basic, label='Predicted Periodic Temperature Model', color='blue')
ax[1].set_title('Seasonal Variation Prediction Error vs. Periodic Temperature Model')
ax[1].set_ylabel('Error (°F)')
ax[1].legend()
ax[1].grid(True)
ax[2].plot(years, error_fourier, label='Prediction error of seasonal variation and enhanced periodic model', color='purple')
ax[2].set_title('Residuals of Seasonal Variation Preduction Error and Error of Enhanced Periodic Temperature Model')
ax[2].set_xlabel('Year')
ax[2].set_ylabel('Error (°F)')
ax[2].legend()
ax[2].grid(True)
plt.show()
```

```python
def linear_regression_fit(x, y):
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
    linear_fit = slope * x + intercept
    return linear_fit, slope, intercept, r_value**2
def exponential_model(x, A, B):
    return A * np.exp(B * x)
def exponential_regression_fit(x, y):
    p_exponential, _ = optimize.curve_fit(exponential_model, x, y, p0=[1, 0.001])
    A, B = p_exponential
    exponential_fit = exponential_model(x, A, B)
    return exponential_fit, A, B

linear_fit_original, slope_original, intercept_original, r_squared_original = linear_regression_fit(time, temperature)
exponential_fit_original, A_original, B_original = exponential_regression_fit(time, temperature)
linear_fit_seasonal, slope_seasonal, intercept_seasonal, r_squared_seasonal = linear_regression_fit(time, error_seasonal)
exponential_fit_seasonal, A_seasonal, B_seasonal = exponential_regression_fit(time, error_seasonal)
linear_fit_fourier, slope_fourier, intercept_fourier, r_squared_fourier = linear_regression_fit(time, error_fourier)
exponential_fit_fourier, A_fourier, B_fourier = exponential_regression_fit(time, error_fourier)

fig, ax = plt.subplots(3, 1, figsize=(15, 15))
ax[0].plot(years, temperature, label='Actual Temperature', color='blue')
ax[0].plot(years, linear_fit_original, label='Linear Fit', linestyle='--', color='green')
ax[0].plot(years, exponential_fit_original, label='Exponential Fit', linestyle='-.', color='orange')
ax[0].set_title('Average Daily Temperature (°F)')
ax[0].set_ylabel('Temperature (°F)')
ax[0].legend()
ax[0].grid(True)
ax[1].plot(years, error_seasonal, label='Prediction Error (Residuals)', color='green')
ax[1].plot(years, linear_fit_seasonal, label='Linear Regression Fit', linestyle='--', color='red')
ax[1].plot(years, exponential_fit_seasonal, label='Exponential Regression Fit', linestyle='-.', color='blue')
ax[1].set_title('Seasonal Variation Prediction Error vs. Periodic Temperature Model')
ax[1].set_ylabel('Error (°F)')
ax[1].legend()
ax[1].grid(True)
ax[2].plot(years, error_fourier, label='Residuals of Fourier Model', color='purple')
ax[2].plot(years, linear_fit_fourier, label='Linear Regression Fit', linestyle='--', color='red')
ax[2].plot(years, exponential_fit_fourier, label='Exponential Regression Fit', linestyle='-.', color='blue')
ax[2].set_title('Residuals of Seasonal Variation Preduction Error and Error of Enhanced Periodic Temperature Model')
ax[2].set_xlabel('Year')
ax[2].set_ylabel('Error (°F)')
ax[2].legend()
ax[2].grid(True)
plt.show()

print(f"Linear regression of original data: Y = {slope_original:.5f}x + {intercept_original:.5f}, R^2: {r_squared_original:.5f}")
print(f"Exponential regression of original data: Y = {A_original:.5f} * e^({B_original:.5f}x)")
print(f"Linear regression of seasonal variation preduction error: Y = {slope_seasonal:.5f}x + {intercept_seasonal:.5f}, R^2: {r_squared_seasonal:.5f}")
print(f"Exponential regression of seasonal variation prediction error: Y = {A_seasonal:.5f} * e^({B_seasonal:.5f}x)")
print(f"Linear regression of residuals graph: Y = {slope_fourier:.5f}x + {intercept_fourier:.5f}, R^2: {r_squared_fourier:.5f}")
print(f"Exponential regression (Fourier Residual): Y = {A_fourier:.5f} * e^({B_fourier:.5f}x)")
```
